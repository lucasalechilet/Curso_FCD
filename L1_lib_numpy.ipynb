#1. Carga y estructuración de datos

```python
import numpy as np
np.random.seed(42) # seteamos el seed en 42 para reproducibilidad


precios_acciones = np.random.uniform(low=500, high=1500, size=(5, 5)) #array de 5x5 con datos uniformes, entre 500 y 1500
precios_acciones = np.round(precios_acciones, 1) #redondeamos a 1 decimal

print("Precios de Acciones")
print(precios_acciones)
print("\n")

```

2. Análisis y Transformación de Datos

```python
promedio_por_accion = np.mean(precios_acciones, axis=1)  #usando numpy, basta usar estos comandos para promediar y obtener min y max
maximo_por_accion = np.max(precios_acciones, axis=1)      #según corresponda. usamos axis=1 para variar sobre día
minimo_por_accion = np.min(precios_acciones, axis=1)

print(f"Promedio: {np.round(promedio_por_accion, 2)}")
print(f"Máximo: {maximo_por_accion}")
print(f"Mínimo: {minimo_por_accion}\n")

variacion_diaria = (precios_acciones[:, 1:] - precios_acciones[:, :-1]) / precios_acciones[:, :-1] * 100  # para calcular variacion porcentual diaria usamos
                                                                                                          #(Precio de hoy - Precio de ayer) / Precio de ayer * 100
                                                                                                          # [:,1:] significa tomar todas las filas, y de las columnas
                                                                                                          #ignoramos el primer valor
                                                                                                          # [:, :-1] significa tomar todas las filas, y de las columnas
                                                                                                          #tomar todos menos el último valor
                                                                                                          #así, podemos aplicar la operación correspondiente para la variación
print("Variación Porcentual Diaria (%)")
print(np.round(variacion_diaria, 1))
print("\n")

#calcularemos normalización
min_dims = minimo_por_accion.reshape(-1, 1)  #reajustamos ambos vectores para aplicar el cálculo a continuación
max_dims = maximo_por_accion.reshape(-1, 1)
precios_normalizados = (precios_acciones - min_dims) / (max_dims - min_dims) #fórmula utilizada: (X - X_min) / (X_max - X_min)

print("Precios Normalizados")
print(np.round(precios_normalizados, 1))
print("\n")
```

3. Optimización y Selección de Datos

```python
#buscamos variación diaria de acción 5 en día 3
rendimiento_especifico = variacion_diaria[4, 1]

precios_altos = precios_acciones[precios_acciones > 1000]

print(f"Precio de la acción 5 en el día 3: ${rendimiento_especifico}")
print(f"Precios totales que superaron los $100: {precios_altos}")
```
