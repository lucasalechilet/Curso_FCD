0. Generación de archivos csv y xslx

```python
import pandas as pd
import numpy as np
import seaborn as sns

df_base = sns.load_dataset('titanic') #cargamos dataset titanic

df_base.iloc[:400].to_csv('pasajeros_p1.csv', index=False) # guardamos los primeros 400 registros en un archivo CSV

df_base.iloc[400:].to_excel('pasajeros_p2.xlsx', index=False, sheet_name='Datos') #guardamos el resto en un archivo Excel

print("1. Archivos de prueba creados\n")
```

1. Carga de datos desde distintos archivos

```python

df_csv = pd.read_csv('pasajeros_p1.csv') #cargamos el csv

df_excel = pd.read_excel('pasajeros_p2.xlsx', sheet_name='Datos') #cargamos el xlsx

url = 'https://es.wikipedia.org/wiki/Anexo:Pasajeros_a_bordo_del_RMS_Titanic' #definimos la url a utilizar

df_web = pd.read_html(url, storage_options={'User-Agent': 'Mozilla/5.0'})[0] # pd.read_html escanea la página y devuelve una lista de todas las tablas que encuentra.
                              # la definición de storaga_options es para simular un usuario y bypassear así el bloqueo de bots.
print(df_web)                 # tomamos la tabla [0], que ene este caso es el de pasajeros de 1a clase

# Unimos las dos fuentes (CSV y Excel) en un solo DataFrame consolidado
df_principal = pd.concat([df_csv, df_excel], ignore_index=True)
```

2. Limpieza y estructuración de datos

```python
df_principal['age'] = df_principal['age'].fillna(df_principal['age'].median()) #rellenamos las edades nulas (columna 'age') con la mediana


df_principal = df_principal.dropna(subset=['embarked']) #eliminamos filas donde falte el dato del puerto de embarque ('embarked')

df_principal = df_principal.drop_duplicates() #eliminamos filas duplicadas (si las hubiera)

# convertimos 'survived' a tipo booleano y 'class' a categoría para ahorrar memoria
df_principal['survived'] = df_principal['survived'].astype(bool)
df_principal['class'] = df_principal['class'].astype('category')
```

3. Transformación y optimización de datos

```python
# seleccionamoscolumnas relevantes
columnas_relevantes = ['survived', 'pclass', 'sex', 'age', 'fare', 'class']
df_optimizado = df_principal[columnas_relevantes]

#renombramos columnas al español para mejor legibilidad
df_optimizado = df_optimizado.rename(columns={
    'survived': 'sobrevivio',
    'pclass': 'clase_ticket',
    'sex': 'sexo',
    'age': 'edad',
    'fare': 'tarifa_pagada',
    'class': 'categoria_clase'
})

#ordenamos los datos por tarifa (de mayor a menor)
df_optimizado = df_optimizado.sort_values(by='tarifa_pagada', ascending=False)
```

4. Exportación de Datos

```python
df_optimizado.to_excel('titanic_dashboard.xlsx', index=False, sheet_name='Datos_Limpios')
```

Conclusiones

Un proceso estructurado de obtención y limpieza de datos (comúnmente llamado ETL: Extract, Transform, Load) es lo más importante de cualquier análisis.“Garbage in, garbage out” es siempre relevante, puesto que ningún modelo estadístico o algoritmo de machine learning, por avanzado que sea, podrá generar conclusiones útiles si se alimenta de datos basura.

La librería Pandas centraliza todo este flujo en un solo lugar. Permite automatizar la unificación de fuentes diversas, garantizando la integridad, la calidad y la trazabilidad de la información antes de entregarla a los tomadores de decisiones.
