# Análisis de Caso: Preprocesamiento y Escalamiento de Datos


---

## 1. Creación del Dataset

```python
#link a github https://github.com/lucasalechilet/Curso_FCD/blob/main/preprocesamiento_y_escalamiento_de_datos.ipynb

import pandas as pd
import numpy as np

data = {
    'ID': [1, 2, 3, 4],
    'Edad': [25, 45, 30, 40],
    'Ciudad': ['Madrid', 'Sevilla', 'Madrid', 'Barcelona'],
    'Ingresos': [30000, 50000, np.nan, 40000]
}

df = pd.DataFrame(data)  #se genera el dataframe con los datos iniciales
df
```

## 2. Imputación de valores faltantes

```python
mean_ingresos = df['Ingresos'].mean()
df['Ingresos'] = df['Ingresos'].fillna(mean_ingresos) #usamos fillna para llenar con la media de ingresos calculada

print("media usada para imputación:", mean_ingresos)
df
```

## 3. Codificación de variables categóricas

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Ciudad_Label'] = le.fit_transform(df['Ciudad']) #aplicamos labelencoder, lo que permitirá codificar cada ciudad como número entero >=0

df
```

### One-Hot Encoding

```python
df_onehot = pd.get_dummies(df, columns=['Ciudad']) #aplicamos onehotencoder a ciudad, lo que permitirá tener un label booleano para cada ciudad, en lugar de 1 sóla columna con string
df_onehot
```

### Variables Dummy (drop_first=True)

```python
df_dummy = pd.get_dummies(df, columns=['Ciudad'], drop_first=True) #aplicamos get_dummies con drop_first=True, lo que eliminará
df_dummy                                                            #una de las columnas implícitas de ciudad. esto se hace para evitar colinealidad,
                                                                    # pues Ciudad_Barcelona queda implícita cuando las otras son False
```

## 4. Escalamiento de variables numéricas

### Normalización Min-Max

```python
from sklearn.preprocessing import MinMaxScaler

scaler_minmax = MinMaxScaler()
df[['Edad_MinMax', 'Ingresos_MinMax']] = scaler_minmax.fit_transform(df[['Edad', 'Ingresos']]) #normalizamos edad e ingresos con min-max

df
```

### Estandarización Z-Score

```python
from sklearn.preprocessing import StandardScaler

scaler_std = StandardScaler()
df[['Edad_Z', 'Ingresos_Z']] = scaler_std.fit_transform(df[['Edad', 'Ingresos']]) #estandarizamos edad e ingresos con método z-score

df
```

## 5. Exportación del dataset preprocesado

```python
df.to_csv('dataset_preprocesado.csv', index=False)
print("Archivo guardado como dataset_preprocesado.csv")
```

## 6. Conclusiones
Realizar estas tareas es fundamental porque la calidad de los datos impacta directamente en el desempeño del modelo de Machine Learning. El preprocesamiento permite que los algoritmos procesen variables categóricas al convertirlas a un formato numérico y evita que valores nulos o escalas muy distintas entre variables generen sesgos o errores en el aprendizaje.

En cuanto a las técnicas de escalamiento, la principal diferencia es que la normalización Min-Max transforma los datos a un rango específico, lo cual es muy útil cuando los datos no siguen una distribución normal. Por otro lado, la estandarización Z-Score reescala los datos para que tengan una media de cero y una desviación estándar de uno, siendo más robusta ante la presencia de valores atípicos y preferible para algoritmos que asumen distribuciones gaussianas.
